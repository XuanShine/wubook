```python
from wubook_api import Wubook
```

```python
dfrom = "19/12/2025"  # input("Date d√©but (DD/MM/YYYY) ? ")
dto = "22/12/2025"  # input("Date fin (DD/MM/YYYY) ? ")
```

```python
w = Wubook()
```

```python
ROOMS = w.fetch_rooms()[1]
ROOMS
ID_TO_ROOMS = {x["id"]: x["name"] for x in ROOMS}
ID_TO_ROOMS
```

```python
ROOMS
```

```python
SUBROOMS = {x["id"]: x["subroom"] for x in ROOMS if x["subroom"]}
SUBROOMS
```

```python
res = w.fetch_rooms_values(dfrom, dto)
if res[0] == 0:
    res = res[1]

res
```

```python
resultat = dict()
for key, value in res.items():
    tmp = {"name": ID_TO_ROOMS[int(key)]}
    for date in value:
        tmp["price"] = tmp.get("price", 0) + date["price"]
        tmp["avail"] = min(tmp.get("avail", 50), date.get("avail", 0))
        tmp["closed"] = max(tmp.get("closed", 0), date["closed"])
        tmp["min_stay"] = max(tmp.get("min_stay", 0), date["min_stay"])
    tmp["price"] = round(tmp["price"])
    resultat[key] = tmp
resultat
```

```python
for key, data in resultat.items():
    if int(key) in SUBROOMS:
        data["avail"] = resultat[str(SUBROOMS[int(key)])]["avail"]
resultat
```
